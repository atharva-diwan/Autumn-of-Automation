```python
import cv2
import numpy as np
```

```python

capture = cv2.VideoCapture(".\Lionel Messi Video.mp4")
 
while(True):
     
    ret, frame = capture.read()
    img_bw=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    ret,thresh = cv2.threshold(img_bw,250,255,0)
    image, contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    for cnt in contours:
        epsilon = 0.005*cv2.arcLength(cnt,True)
        approx = cv2.approxPolyDP(cnt,epsilon,True)
        cv2.drawContours(img_bw, [approx], 0, (0), 5)
        x = approx.ravel()[0]
        y=approx.ravel()[1]
        if len(approx)>14:
            cv2.putText(img_bw, "football", (x, y), cv2.FONT_HERSHEY_COMPLEX, 1, (0))
    cv2.imshow('video', img_bw)
     
    if cv2.waitKey(120) & 0xFF == ord('q'):
        break
 
capture.release()
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

MIN_MATCH_COUNT = 2
capture = cv2.VideoCapture(".\Lionel Messi Video.mp4")
while(True):
     
    ret, frame = capture.read()
              # queryImage
    im=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imwrite('.\img2.jpg',im)
    img1 = cv2.imread('foot_query.jpg',0)          # queryImage
    img2 = cv2.imread('.\img2.jpg',0) # trainImage

    # Initiate SIFT detector
    #orb = cv2.ORB_create()   
    sift = cv2.xfeatures2d.SIFT_create()
    # find the keypoints and descriptors with SIFT
    kp1, des1 = sift.detectAndCompute(img1,None)
    kp2, des2 = sift.detectAndCompute(img2,None)

    FLANN_INDEX_KDTREE = 0
    index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 3)
    search_params = dict(checks = 50)

    flann = cv2.FlannBasedMatcher(index_params, search_params)
    try:
        matches = flann.knnMatch(des1,des2,k=2)
    except:
        continue
    good = []
    for m,n in matches:
        if m.distance < 0.6*n.distance:
            good.append(m)
    if len(good)>MIN_MATCH_COUNT:
        src_pts = np.float32([ kp1[m.queryIdx].pt for m in good ]).reshape(-1,1,2)
        dst_pts = np.float32([ kp2[m.trainIdx].pt for m in good ]).reshape(-1,1,2)

        M, mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC,5.0)
        matchesMask = mask.ravel().tolist()

        h,w = img1.shape
        pts = np.float32([ [0,0],[0,h-1],[w-1,h-1],[w-1,0] ]).reshape(-1,1,2)
        dst = cv2.perspectiveTransform(pts,M)

        img2 = cv2.polylines(img2,[np.int32(dst)],True,255,3, cv2.LINE_AA)

    else:
        print("Not enough matches are found - %d/%d" % (len(good),MIN_MATCH_COUNT))
        matchesMask = None
    draw_params = dict(matchColor = (0,255,0), # draw matches in green color
                       singlePointColor = None,
                       matchesMask = matchesMask, # draw only inliers
                       flags = 2)

    img3 = cv2.drawMatches(img1,kp1,img2,kp2,good,None,**draw_params)
    cv2.imshow('video', img3)
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
 
capture.release()
cv2.destroyAllWindows()
```
